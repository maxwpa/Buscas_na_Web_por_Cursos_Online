```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cursos = pd.read_csv('Buscas_por_Cursos.csv', skiprows=1, parse_dates=['Semana'], index_col='Semana')
ead = pd.read_csv('Buscas_por_Cursos_EAD.csv', skiprows=1, parse_dates=['Semana'], index_col='Semana')
online = pd.read_csv('Buscas_por_Cursos_Online.csv', skiprows=1, parse_dates=['Semana'], index_col='Semana')
```

```python
print(cursos.columns)
print(ead.columns)
print(online.columns)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(cursos.index, cursos['cursos: (Brasil)'], label='Cursos', color='blue')
plt.plot(ead.index, ead['cursos ead: (Brasil)'], label='Cursos EAD', color='orange')
plt.plot(online.index, online['cursos online: (Brasil)'], label='Cursos Online', color='green')
plt.title('Busca por Cursos na Web (2021 - 2023)', fontsize='20', fontweight='bold')
plt.legend()
plt.grid()
plt.show()
```
