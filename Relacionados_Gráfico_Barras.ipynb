```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nivel = pd.read_excel('Palavras_mais_Relacionadas_Cursos.xlsx')
```

```python
nivel.columns
```

```python
nivel = nivel.sort_values(by='NÃ­vel Relativo')
fig, axs = plt.subplots(figsize=(10, 6))
axs.barh(nivel['Palavras-Chaves'], nivel['NÃ­vel Relativo'], color='purple')
axs.set_xlabel('')
axs.set_ylabel('')
axs.set_title('Palavras Relacionadas mais Pesquisadas em 2023', fontsize='20', fontweight='bold')
axs.grid(True)
plt.tight_layout()
plt.show()
```
