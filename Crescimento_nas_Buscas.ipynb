```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taxa = pd.read_excel('Palavras_Relacionada_Maior_Crescimento.xlsx')
```

```python
taxa = taxa.sort_values(by='Taxa de Crescimento')
fig, axs = plt.subplots(figsize=(12, 6))
bars = axs.barh(taxa['Palavras-Chaves'], taxa['Taxa de Crescimento'], color='red')
axs.set_xlabel('')
axs.set_ylabel('')
axs.set_title('Taxa de Crecimento das Palavras Pesquisadas à Cursos em 2023', fontsize='20', fontweight='bold')
axs.grid(True)
for bar in bars:
    width = bar.get_width()
    axs.text(width, bar.get_y() + bar.get_height()/2, f'{int(width)}%', va='center')
plt.tight_layout()
plt.show()
```
